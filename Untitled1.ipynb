Relação linear de salário

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv'
df = pd.read_csv(url)

print("Dados importados com sucesso. As 5 primeiras linhas são:")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.regplot(x='Experience Years', y='Salary', data=df)
plt.title('Relação Linear entre Anos de Experiência e Salário')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Preparar os dados para o modelo de regressão linear
X = df[['Experience Years']].values # Variável independente (anos de experiência)
y = df['Salary'].values           # Variável dependente (salário)

# Criar e ajustar o modelo de regressão linear
model = LinearRegression()
model.fit(X, y)

# Obter o coeficiente angular (slope) e o coeficiente linear (intercept)
coeficiente_angular = model.coef_[0]
coeficiente_linear = model.intercept_

print(f'Coeficiente Angular (Inclinação): {coeficiente_angular:.2f}')
print(f'Coeficiente Linear (Intercepto): {coeficiente_linear:.2f}')
```
